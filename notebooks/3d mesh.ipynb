```julia
# Loading the finite element package Amaru
using Revise
using FemMesh
```

```julia
# Mesh generation
blocks2d = [
          Block2D([0 0; 0.3 2], nx=4, ny=20, cellshape=QUAD4 ),
          Block2D([0 2; 0.3 2.35], nx=4, ny=4, cellshape=QUAD4 ),
          Block2D([0 2.35; 0.3 2.65], nx=4, ny=4, cellshape=QUAD4 ),
          Block2D([0.3 2; 0.55 2.15; 0.55 2.35; 0.3 2.35], nx=4, ny=4, cellshape=QUAD4 ),
         ]
;
```

```julia
mplot(blocks2d)
```

```julia
blocks = extrude(blocks2d, len=0.2, n=4)
;
@show length(blocks)
```

```julia
mplot(blocks)
```

```julia
roll_axes!(blocks) # x y z -> y z x
```

```julia
mplot(blocks)
```

```julia
bl = BlockInset([0.03 0.03 0.03; 0.03 0.03 2.62], cellshape=LIN3, tag="lbars")
lbars = array(bl, nx=2, ny=2, dx=0.14, dy=0.24)

bl = BlockInset([0.04 0.03 1.78; 0.04 0.03 1.9; 0.04 0.52 2.17; 0.04 0.52 2.32; 0.04 0.03 2.32; 0.04 0.03 2.15], cellshape=LIN3, tag="cbars")
cbars = array(bl, nx=2, dx=0.12)

bl = BlockInset([0.03 0.03 0.15; 0.17 0.03 0.15; 0.17 0.27 0.15; 0.03 0.27 0.15; 0.03 0.03 0.15], cellshape=LIN3, tag="sbars1")
sbars1 = array(bl, nz=13, dz=0.2)

bl = BlockInset([0.03 0.03 2.2; 0.17 0.03 2.2; 0.17 0.52 2.2; 0.03 0.52 2.2; 0.03 0.03 2.2], cellshape=LIN3, tag="sbars2")
sbars2 = array(bl, nz=3, dz=0.05)
;
```

```julia
mplot(blocks, lbars, cbars, sbars1, sbars2, azim=30)
```

```julia
msh = Mesh(blocks, lbars, cbars, sbars1, sbars2)
;
```

```julia
mplot(msh, field="point-id", azim=30)
```

```julia
mplot(msh, field="quality", azim=30)
```
